```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
### filepaths go here 
### csv with first colum as q-values and following columns are integrations
### top row of csv should be film thickness in nm

## blend RT OoP
#filepath = r"/home/goughl/Documents/EA_paper_working/EA_blend_RT/all_1d_int_EA_blend_OoP.csv"
#filepath = r"/home/goughl/Documents/EA_paper_working/blend_RT/all_1d_int_blend_OoP.csv"

## neat HT OoP
#filepath = r"/home/goughl/Documents/EA_paper_working/EA_DCV5T_HT_ver2_long_exposure/all_1d_int_EA_DCV5T_HT_OoP.csv"
filepath = r"/home/goughl/Documents/EA_paper_working/DCV5T_HT_ver2_long_exposure/all_1d_int_DCV5T_HT_OoP.csv"

#filepath = r"/home/goughl/Documents/EA_paper_working/EA_RT/all_1d_int_EA_RT_OoP.csv"

## neat RT OoP
#filepath = r"/home/goughl/Documents/EA_paper_working/DCV5T_RT_ver2/all_1d_int_DCV5T_RT_ver2_OoP.csv"

df = pd.read_csv(filepath, skiprows=[1])

# set x-axis colunm as the index (q-values)
df.set_index(df.columns[0], inplace=True)

# get the background column (first dataset column, first integration in an in situ run)
background = df.iloc[:, 0]

# subtract the background from all datasets
adjusted_df = df.sub(background, axis=0)
```

```python
### plotting the data with no manipulation 

plt.figure(figsize=(10, 6))

for column in adjusted_df.columns:
    plt.plot(adjusted_df.index, df[column], label=column)

plt.xlabel("Intensity (a.u.)")
plt.ylabel("Q (1/Å)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
### plot data with first image of in situ run subtracted as background
plt.figure(figsize=(10, 6))

for column in adjusted_df.columns:
    plt.plot(adjusted_df.index, adjusted_df[column], label=column)

plt.xlabel("Intensity (a.u.)")
plt.ylabel("Q (1/Å)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
### customisable heatmap

# can shift all the datasets by an increment here
shifted_df = adjusted_df + 0.02

# to plot only a certain range of q values change these two numbers and uncomment
#shifted_df = adjusted_df[(adjusted_df.index >= 0.5) & (adjusted_df.index <= 1.25)]

# select only the range of datasets between 0 - 50 nm
shifted_df_20 = shifted_df.iloc[:, 2:26]
plt.figure(figsize=(8, 6))

# change vmin and vmax to change the scale range
sns.heatmap(shifted_df_20.T, cmap="viridis", cbar_kws={'label': 'Intensity Scale (a.u.)'},
            vmin=0, vmax=0.075)

plt.xlabel("Q (1/Å)")
plt.ylabel("Film thickness (nm)")
plt.title("Integration out of plane as a function of film thickness")

x_values = shifted_df.index 
plt.xticks(
    np.arange(0, len(x_values), int(0.1 / (x_values[1] - x_values[0]))),
    labels=np.round(x_values[::int(0.1 / (x_values[1] - x_values[0]))], 1),
    rotation=45
)

plt.show()
```
